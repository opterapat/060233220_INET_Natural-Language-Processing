## 2.1 `.split()` — การแบ่งข้อความเป็น Tokens

`.split()` คือ **Whitespace Tokenizer** ที่ง่ายที่สุด

````python
# ── พื้นฐาน .split() ────────────────────────────────────────
sentence = "The quick brown fox jumps over the lazy dog"

# แบ่งด้วย whitespace (default)
tokens = sentence.split()
print("tokens:", tokens)
print("จำนวน tokens:", len(tokens))
print()

# แบ่งด้วยตัวคั่นที่กำหนดเอง
csv_line = "Bangkok,Chiang Mai,Phuket,Pattaya"
cities = csv_line.split(",")
print("cities:", cities)
print()

# จำกัดจำนวนการแบ่ง
text = "one two three four five"
print("split(maxsplit=3):", text.split(maxsplit=3))
````

````python
# ── .splitlines() สำหรับข้อความหลายบรรทัด ──────────────────
paragraph = """NLP คือการทำให้คอมพิวเตอร์เข้าใจภาษา
มันใช้ในการแปลภาษา การสรุปข้อความ
และการวิเคราะห์ความรู้สึก"""

lines = paragraph.splitlines()
print("จำนวนบรรทัด:", len(lines))
for i, line in enumerate(lines, 1):
    print(f"  บรรทัด {i}: '{line}'")
print()

# ── .join() — รวม list เป็น string ───────────────
tokens = ["The", "cat", "sat", "on", "the", "mat"]
sentence = " ".join(tokens)
print("join ด้วยช่องว่าง :", sentence)

# กรณีหลังการทำ tokenize แล้วต้องการ rejoin
processed = [w.lower() for w in tokens if w.lower() not in ["the", "on"]]
print("หลังลบ stopwords:", " ".join(processed))
````

## 2.2 `.strip()`, `.lstrip()`, `.rstrip()` — การกำจัด Whitespace

ในบางครั้งข้อมูลดิบที่ได้มา (web scraping, CSV files) มักมี whitespace ที่ไม่ต้องการ

````python
# ── .strip() พื้นฐาน ─────────────────────────────────────────
messy = "   Hello, World!   "
print(f"ก่อน strip : |{messy}|")
print(f".strip()   : |{messy.strip()}|")
print(f".lstrip()  : |{messy.lstrip()}|")
print(f".rstrip()  : |{messy.rstrip()}|")
print()

# ── strip ตัวอักษรที่กำหนด ───────────────────────────────────
url = "###Important Text###"
print(f"strip('#') : {url.strip('#')}")

punctuated = "...Hello World..."
print(f"strip('.'): {punctuated.strip('.')}")
print()

# Use case: ทำความสะอาดข้อมูลจาก CSV
raw_data = ["  Bangkok ", "\tChiang Mai\n", "  Phuket  "]
print("ก่อน clean:", raw_data)
clean_data = [city.strip() for city in raw_data]
print("หลัง clean :", clean_data)
````

## 2.3 `.replace()` — การแทนที่ข้อความ

ใช้สำหรับ **Text Cleaning** — ลบหรือแทนที่ตัวอักษรที่ไม่ต้องการ

````python
# ── .replace() พื้นฐาน ──────────────────────────────────────
text = "I love cats. Cats are great. My cat is named Whiskers."
print("ต้นฉบับ :", text)
print(".replace():", text.replace("cat", "dog"))
print()

# จำกัดจำนวนครั้งที่แทนที่
print("replace 1 ครั้ง:", text.replace("cat", "dog", 1))
print()

# ── Use cases ใน NLP ──────────────────────────────────────
# 1) ลบ HTML tags
html_text = "<p>Hello <b>World</b>!</p>"
cleaned = html_text.replace("<p>", "").replace("</p>", "").replace("<b>", "").replace("</b>", "")
print("ลบ HTML tags:", cleaned)

# 2) Normalize punctuation
noisy = "Hello!!!   How are you???   Fine..."
normalized = noisy.replace("!!!", "!").replace("???", "?").replace("...", ".")
print("normalize punct:", normalized)

# 3) ลบ newlines
multiline = "First line\nSecond line\nThird line"
single = multiline.replace("\n", " ")
print("รวมบรรทัด:", single)
````

## 2.4 `.find()`, `.count()`, `.startswith()`, `.endswith()` — การค้นหาใน String

````python
text = "NLP stands for Natural Language Processing. NLP is powerful!"

# ── .find() และ .index() ─────────────────────────────────────
print("=== .find() ===")
pos = text.find("NLP")
print(f"'NLP' พบครั้งแรกที่ตำแหน่ง: {pos}")
print(f"'NLP' พบครั้งที่ 2 ที่ตำแหน่ง: {text.find('NLP', pos+1)}")
print(f"'xyz' พบที่: {text.find('xyz')}  (ไม่พบ = -1)")
print()

# ── .count() ─────────────────────────────────────────────────
print("=== .count() ===")
print(f"'NLP' ปรากฏ {text.count('NLP')} ครั้ง")
print(f"'a' ปรากฏ {text.count('a')} ครั้ง")
print()

# ── .startswith() และ .endswith() ───────────────────────────
print("=== .startswith() / .endswith() ===")
sentences = [
    "Hello world",
    "NLP is fun",
    "Machine learning rocks",
    "Natural Language Processing"
]
nlp_sentences = [s for s in sentences if s.startswith("NLP") or s.startswith("Natural")]
print("ประโยคที่เริ่มต้นด้วย NLP/Natural:", nlp_sentences)

# ── 'in' operator ────────────────────────────────────────────
print()
print("=== 'in' operator ===")
print(f"'Natural' in text: {'Natural' in text}")
print(f"'Python' in text: {'Python' in text}")
````

````python
#เขียนโค้ดของคุณที่นี่
def clean_tweet(text):
    """
    ทำความสะอาดข้อความ Twitter
    Args:
        text (str): ข้อความ tweet
    Returns:
        list: list ของ tokens หลังทำความสะอาด
    """
    text = text.lower()  # ทำให้เป็นตัวพิมพ์เล็ก
    text = text.strip()  # ลบช่องว่างที่อยู่ด้านหน้าและด้านหลัง
    tokens = text.split()  # แบ่งข้อความเป็น tokens
    cleaned_tokens = []
    for token in tokens:
        # ลบ mentions, hashtags, และ URLs
        if token.startswith('@') or token.startswith('#') or token.startswith('http'):
            continue
        cleaned_tokens.append(token)
    return cleaned_tokens
   
    pass


# Test cases
tweets = [
    "Just learned about @OpenAI #ChatGPT - it's amazing! https://openai.com",
    "  #NLP is the future! Check out https://huggingface.co @huggingface  ",
    "@elonmusk said something about #AI today https://twitter.com/xyz",
]

print("=== ทดสอบ clean_tweet() ===")
for tweet in tweets:
    print(f"Input : {tweet}")
    result = clean_tweet(tweet)
    print(f"Output: {result}")
    print()
````

# Part 3: Unicode & การจัดการข้อความหลายภาษา

Unicode คือมาตรฐานสากลสำหรับการเข้ารหัสข้อความ — รองรับทุกภาษาในโลก  
ความเข้าใจเรื่อง Unicode เป็นสิ่งจำเป็นสำหรับ **Multilingual NLP** (การประมวลผลภาษาธรรมชาติหลายภาษา)

## 3.1 Unicode คืออะไร?

```
ตัวอักษร  →  Code Point  →  Encoding (bytes)
   'A'    →   U+0041     →  0x41 (UTF-8: 1 byte)
   'ก'    →   U+0E01     →  0xE0 0xB8 0x81 (UTF-8: 3 bytes)
   '🤖'   →   U+1F916    →  0xF0 0x9F 0xA4 0x96 (UTF-8: 4 bytes)
```

````python
# ── Code Points ──────────────────────────────────────────────
chars = ['A', 'a', 'ก', 'ข', '🤖', '😊']

print(f"{'ตัวอักษร':<8} {'Code Point':<12} {'ค่าทศนิยม':<12} {'ชื่อ Unicode'}")
print("-" * 60)
import unicodedata
for ch in chars:
    cp = ord(ch)
    try:
        name = unicodedata.name(ch)
    except ValueError:
        name = "(emoji)"
    print(f"  {ch:<7} U+{cp:04X}      {cp:<12} {name}")
````

````python
# ── Encoding: แปลง String → Bytes ───────────────────────────
text = "Hello ภาษาไทย"

print("=== Encoding Comparison ===")
for encoding in ['utf-8', 'utf-16', 'ascii']:
    try:
        encoded = text.encode(encoding)
        print(f"{encoding:8}: {len(encoded)} bytes → {encoded[:20]}{'...' if len(encoded)>20 else ''}")
    except UnicodeEncodeError as e:
        print(f"{encoding:8}: Error - {e}")
print()

# ── Decoding: แปลง Bytes → String ───────────────────────────
print("=== Decoding ===")
utf8_bytes = text.encode('utf-8')
decoded = utf8_bytes.decode('utf-8')
print(f"Original : {text}")
print(f"Encoded  : {utf8_bytes}")
print(f"Decoded  : {decoded}")
print(f"Match    : {text == decoded}")
````

````python
# ── unicodedata: ข้อมูลเพิ่มเติมของ Unicode ─────────────────
import unicodedata

test_chars = ['A', 'ก', '1', ' ', '!', '😊']
print(f"{'Char':<6} {'Category':<10} {'Is Alpha':<10} {'Is Digit':<10} {'Is Space'}")
print("-" * 55)
for ch in test_chars:
    cat = unicodedata.category(ch)
    print(f"  {ch!r:<5} {cat:<10} {str(ch.isalpha()):<10} {str(ch.isdigit()):<10} {ch.isspace()}")
print()

# ── Unicode Normalization ────────────────────────────────────
# ปัญหา: 'é' (café) สามารถเขียนได้ 2 แบบใน Unicode
e_combined  = 'é'          # U+00E9 (1 code point)
e_decomposed = 'e\u0301'   # e + combining accent (2 code points)

print("=== Unicode Normalization ===")
print(f"e_combined  : '{e_combined}'  len={len(e_combined)}")
print(f"e_decomposed: '{e_decomposed}'  len={len(e_decomposed)}")
print(f"เท่ากันไหม? {e_combined == e_decomposed}")

# แก้ด้วย NFC normalization
nfc_combined  = unicodedata.normalize('NFC',  e_combined)
nfc_decomposed = unicodedata.normalize('NFC', e_decomposed)
print(f"หลัง NFC normalize: {nfc_combined == nfc_decomposed}")  # True!
````


Note:

*   **Lu (Letter, uppercase)**: ตัวอักษรตัวพิมพ์ใหญ่ เช่น 'A', 'B', 'Z'
*   **Lo (Letter, other)**: ตัวอักษรประเภทอื่นๆ ที่ไม่ใช่ตัวพิมพ์ใหญ่หรือเล็กของภาษาละติน มักจะเป็นตัวอักษรของภาษาอื่น เช่น 'ก' (ภาษาไทย)
*   **Nd (Number, decimal digit)**: ตัวเลขฐานสิบ เช่น '0', '1', '9'
*   **Zs (Separator, space)**: ช่องว่างต่างๆ เช่น ' ' (space), '\t' (tab)
*   **Po (Punctuation, other)**: เครื่องหมายวรรคตอนอื่นๆ ที่ไม่ใช่เครื่องหมายเปิด-ปิดวงเล็บหรือเครื่องหมายขีดคั่น เช่น '!', '?', ','
*   **So (Symbol, other)**: สัญลักษณ์อื่นๆ เช่น '😊' (อีโมจิ), '™' (เครื่องหมายการค้า)

NFC

NFC ย่อมาจาก Normalization Form C (Composed) เป็นรูปแบบการทำให้เป็นมาตรฐาน (Normalization Form) ของ Unicode

บทบาทสำคัญของมันคือการจัดการกับตัวอักษรบางตัวที่สามารถเขียนแทนได้หลายแบบใน Unicode (เช่น ตัวอักษรที่มีเครื่องหมายกำกับอย่าง 'é' ซึ่งอาจเขียนเป็นอักษรเดี่ยว (precomposed) หรือเป็นอักษร 'e' รวมกับเครื่องหมายกำกับ (combining accent) ก็ได้) NFC จะแปลงทุกรูปแบบให้เป็นรูปแบบที่เป็น 'composed' หรือรวมกันเป็นอักษรเดียวเสมอ

NFD

NFD ย่อมาจาก Normalization Form D (Decomposed) เป็นอีกรูปแบบหนึ่งของการทำให้เป็นมาตรฐาน (Normalization Form) ใน Unicode

NFD ทำหน้าที่ตรงกันข้ามกับ NFC คือ แทนที่จะรวมตัวอักษรที่มีเครื่องหมายกำกับให้เป็นรหัสเดียว (composed form) NFD จะแยกตัวอักษรนั้นออกเป็นตัวอักษรหลักและเครื่องหมายกำกับ (combining marks) อย่างชัดเจน (decomposed form)

ตัวอย่างเช่น ตัวอักษร 'é' (U+00E9) ในรูปแบบ NFC จะถูกคงไว้เป็นรหัสเดียว แต่ในรูปแบบ NFD มันจะถูกแยกออกเป็น 'e' (U+0065) ตามด้วยเครื่องหมายกำกับเสียงสูง (U+0301). การแยกนี้มีประโยชน์ในบางกรณี เช่น การลบเครื่องหมายกำกับเสียง (accents) ออกจากข้อความ เพื่อทำ Text Normalization ให้เหลือแต่ตัวอักษรหลักในการประมวลผล





````python
# ── Use Case: ทำความสะอาดข้อความหลายภาษา ────────────────
import unicodedata

def remove_accents(text):
    """ลบ accent marks ออกจากข้อความ (สำหรับ language normalization)"""
    # NFD แยก character กับ accent ออกจากกัน
    nfd = unicodedata.normalize('NFD', text)
    # กรองเอาเฉพาะตัวอักษรหลัก (ไม่เอา combining marks)
    return ''.join(ch for ch in nfd
                   if unicodedata.category(ch) != 'Mn')

def keep_only_letters_and_spaces(text):
    """เก็บเฉพาะตัวอักษรและช่องว่าง (ทุกภาษา)"""
    return ''.join(ch for ch in text
                   if ch.isalpha() or ch.isspace())

texts = [
    "Héllo Wörld!",
    "Café au lait 123",
    "ภาษาไทย NLP!",
]

print(f"{'ต้นฉบับ':<25} {'ลบ accents':<25} {'ตัวอักษรอย่างเดียว'}")
print("-" * 75)
for t in texts:
    no_acc = remove_accents(t)
    letters = keep_only_letters_and_spaces(t)
    print(f"{t:<25} {no_acc:<25} {letters}")
````

````python
import csv

# ── วิธีที่ 2: ใช้ csv.DictReader (แนะนำ) ───────────────────
print("=== csv.DictReader ===")
reviews_data = []
with open("lab_data/reviews.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        reviews_data.append(dict(row))

print(f"โหลดข้อมูลทั้งหมด {len(reviews_data)} รายการ")
print("ตัวอย่าง:", reviews_data[0])
print()

# วิเคราะห์ข้อมูล
from collections import Counter
sentiments = Counter(row['sentiment'] for row in reviews_data)
print("การกระจาย Sentiment:")
for sentiment, count in sentiments.items():
    bar = '█' * count
    print(f"  {sentiment:8}: {bar} ({count})")
````

# Part 5: Mini Project — Simple NLP Text Preprocessor

นำทุกอย่างที่เรียนมารวมกันเพื่อสร้าง **Text Preprocessing Pipeline** ฉบับสมบูรณ์

````python
import re
import json
import unicodedata
from collections import Counter

class TextPreprocessor:
    """
    Simple NLP Text Preprocessor
    รวม String Operations และ Unicode handling ที่เรียนมาทั้งหมด
    """

    def __init__(self, language="en"):
        self.language = language
        self.stopwords_en = {"the", "a", "an", "is", "are", "was", "were",
                             "be", "been", "being", "have", "has", "had",
                             "do", "does", "did", "will", "would", "could",
                             "should", "may", "might", "shall", "can",
                             "to", "of", "in", "for", "on", "with", "at",
                             "by", "from", "it", "its", "this", "that"}

    def lowercase(self, text):
        """แปลงเป็นตัวพิมพ์เล็ก"""
        return text.............()

    def strip_whitespace(self, text):
        """ลบ whitespace ส่วนเกิน"""
        # ลบ whitespace ซ้อนกัน
        text = " ".join(text................())
        return text.strip()

    def remove_urls(self, text):
        """ลบ URL"""
        tokens = text................()
        tokens = [t for t in tokens if not t.startswith(('.....', '.....', 'ftp'))]
        return " ".join(tokens)

    def remove_mentions_hashtags(self, text):
        """ลบ @mentions และ #hashtags"""
        tokens = text.split()
        tokens = [t for t in tokens if not t.startswith(('....', '....'))]
        return " ".join(tokens)

    def normalize_unicode(self, text):
        """ทำ Unicode normalization"""
        return unicodedata.normalize('NFC', text)

    def tokenize(self, text):
        """แบ่งข้อความเป็น tokens"""
        # ลบ punctuation และแบ่งด้วย whitespace
        clean = "".join(ch if (ch.isalnum() or ch.isspace()) else " " for ch in text)
        return [t for t in clean.split() if t]

    def remove_stopwords(self, tokens):
        """ลบ stopwords"""
        return [t for t in tokens if t not in self.stopwords_en]

    def preprocess(self, text, steps=None):
        """รัน preprocessing pipeline"""
        if steps is None:
            steps = ['lowercase', 'normalize_unicode', 'strip_whitespace',
                     'remove_urls', 'remove_mentions_hashtags', 'tokenize']

        log = {"original": text, "steps": []}

        current = text
        for step in steps:
            if step == 'tokenize':
                result = self.tokenize(current)
                log['steps'].append({"step": step, "output": result})
                log['tokens'] = result
                return log
            else:
                method = getattr(self, step)
                current = method(current)
                log['steps'].append({"step": step, "output": current[:80] + '...' if len(current)>80 else current})

        log['tokens'] = current.split()
        return log


print(" TextPreprocessor class พร้อมใช้งาน")
````

````python
# ── ทดสอบ TextPreprocessor ───────────────────────────────────
preprocessor = TextPreprocessor()

test_texts = [
    "  Hello @world! Check out https://nlp.example.com #NLP #AI  ",
    "Natural Language Processing is AMAZING and very POWERFUL!",
    "ภาษาไทย NLP @user #hashtag http://example.com",
]

for text in test_texts:
    print("=" * 60)
    result = preprocessor.preprocess(text)
    print(f"INPUT  : {result['original'].strip()}")
    for step_info in result['steps']:
        print(f"  [{step_info['step']}] → {step_info['output']}")
    print(f"TOKENS : {result['tokens']}")
    print()
````

````python
import csv
from collections import Counter

# ── ประมวลผล CSV Dataset ด้วย TextPreprocessor ───────────────
print("=== ประมวลผล Reviews Dataset ===")
preprocessor = TextPreprocessor()
all_tokens = []
results = []

with open("lab_data/reviews.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        result = preprocessor.preprocess(row['text'])
        tokens = result['tokens']
        all_tokens.extend(tokens)
        results.append({
            'id': row['id'],
            'original': row['text'],
            'tokens': tokens,
            'token_count': len(tokens),
            'sentiment': row['sentiment']
        })

# สรุปผล
print(f"ประมวลผล {len(results)} reviews\n")

word_freq = Counter(all_tokens)
print("Top 10 คำที่พบบ่อยที่สุด:")
for word, count in word_freq.most_common(10):
    bar = '▓' * count
    print(f"  {word:<15}: {bar} ({count})")

# บันทึกผลลัพธ์เป็น JSON
with open("lab_data/preprocessed_results.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)

print(f"\nบันทึกผลลัพธ์ไปยัง preprocessed_results.json")
````
